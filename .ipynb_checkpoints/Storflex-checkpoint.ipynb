# Analysis for Storflex 

## Importing modules

```python
from usgs_api import get_data
from hydrogenerate import calculate_potential
from hydrogenerate_plot import get_plot
```

Identified potential stream gauges are:  <br> 1. Site Number: 14158790
Site Name: SMITH RIVER ABV SMITH R RESV,NR BELKNAP SPRNGS,OR, and <br> 2. Site Number: 14158798
Site Name: SMITH RIVER ABV TRAIL BRDG RESV NR BELKNAP SPRINGS

```python
site_no = ['14158790','14158798']
begin_date = '2020-01-01'  #format: 'YYYY-MM-DD'
end_date = '2020-12-31'
```

```python
values = {"flow_info_"+str(site_no[i]):get_data(site_no[i],begin_date,end_date) for i in range(len(site_no))}
```

```python
calculate_x= {"info_"+str(site_no[t]):calculate_potential(values[i],rated_flow=1165,turb= 'Francis turbine', rated_power= 50,
                       head_input=500,op='Timeseries',system='reservoir')for t,i in enumerate(values)} 
```

## Calculating for 14158790

```python
power_1 = calculate_x['info_14158790']["power"]

efficiency_1 = calculate_x['info_14158790']["efficiency"]

flow_range_1 = calculate_x['info_14158790']["flow"]

turb_cap_1 = calculate_x['info_14158790']["turbine_capacity"]
```

## Plotting for 14158790

```python
get_plot(power_1,efficiency_1,flow_range_1,op='Timeseries',flow_info=values['flow_info_14158790'],rated_flow =1165)
```

## Calculating for 14158798

```python
power_2 = calculate_x['info_14158798']["power"]

efficiency_2 = calculate_x['info_14158798']["efficiency"]

flow_range_2 = calculate_x['info_14158798']["flow"]

turb_cap_2 = calculate_x['info_14158798']["turbine_capacity"]
```

## Plotting for 14158798

```python
get_plot(power_2,efficiency_2,flow_range_2,op='Timeseries',flow_info=values['flow_info_14158798'],rated_flow =1165)
```

```python
import pandas as pd
df=pd.DataFrame.from_dict(calculate_x['info_14158790'])

df.to_csv('Data_calculated_14158790.csv',index=False)
```

```python
df=pd.DataFrame.from_dict(calculate_x['info_14158798'])

df.to_csv('Data_calculated_14158798.csv',index=False)
```
